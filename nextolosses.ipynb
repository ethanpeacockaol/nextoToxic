# Nexto Toxic Losses format my score vs their score
# we will cahnge format if my score gets above theirs 
# if i score their score will just be the detriment of theirs vs mine

```python
losses = []
myscore = []
theirscore = []
mysaves = []
theirsaves = []

losses.append(21)
myscore.append(255)
theirscore.append(2891)
mysaves.append(1)
theirsaves.append(0)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.title('Nexto Loss Stats')
plt.plot([i for i in range(len(losses))], losses, 'r.', label="losses")
plt.plot([i for i in range(len(losses))], losses, 'r')
plt.plot([i for i in range(len(myscore))], myscore, 'b', label="score: user")
plt.plot([i for i in range(len(theirscore))], theirscore, 'b--', label="score: nexto")
plt.plot([i for i in range(len(mysaves))], mysaves, 'k', label="saves: user")
plt.plot([i for i in range(len(theirsaves))], theirsaves, 'k--', label="saves: nexto")
plt.legend()
plt.show()
```
